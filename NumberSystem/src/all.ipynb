```python
def binary_to_decimal(binary_str):
    return int(binary_str, 2)

# Example
binary_num = '1011'  # binary for 11
decimal_num = binary_to_decimal(binary_num)
print(f"Binary {binary_num} to Decimal is: {decimal_num}")
```

```python
def binary_to_octal(binary_str):
    decimal = int(binary_str, 2)  # Convert binary to decimal
    return oct(decimal)[2:]  # Convert decimal to octal and strip '0o' prefix

# Example
binary_num = '1011'  # binary for 11
octal_num = binary_to_octal(binary_num)
print(f"Binary {binary_num} to Octal is: {octal_num}")
```

```python
def decimal_to_binary(decimal_num):
    return bin(decimal_num)[2:]  # Convert decimal to binary and strip '0b' prefix

# Example
decimal_num = 11
binary_num = decimal_to_binary(decimal_num)
print(f"Decimal {decimal_num} to Binary is: {binary_num}")
```

```python
def decimal_to_octal(decimal_num):
    return oct(decimal_num)[2:]  # Convert decimal to octal and strip '0o' prefix

# Example
decimal_num = 11
octal_num = decimal_to_octal(decimal_num)
print(f"Decimal {decimal_num} to Octal is: {octal_num}")
```

```python
def octal_to_binary(octal_str):
    decimal = int(octal_str, 8)  # Convert octal to decimal
    return bin(decimal)[2:]  # Convert decimal to binary and strip '0b' prefix

# Example
octal_num = '13'  # octal for 11
binary_num = octal_to_binary(octal_num)
print(f"Octal {octal_num} to Binary is: {binary_num}")
```

```python
def octal_to_decimal(octal_str):
    return int(octal_str, 8)  # Convert octal to decimal

# Example
octal_num = '13'  # octal for 11
decimal_num = octal_to_decimal(octal_num)
print(f"Octal {octal_num} to Decimal is: {decimal_num}")
```
