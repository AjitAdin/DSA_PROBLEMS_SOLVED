````python
# parametrized 
def reverse(arr,l,r):
    if l>=r:
        return
    arr[l], arr[r] = arr[r], arr[l]
    
    reverse(arr,l+1,r-1)

arr=[1,2,3,4,2]
reverse(arr,0,len(arr)-1)
print(arr)
````

````python
# check if the string is palindrome or not
def is_palindrome_slicing(s):
    s = s.lower().replace(" ", "")  # Optional: ignore case and spaces
    return s == s[::-1]

# Example
print(is_palindrome_slicing("Madam"))  # True
print(is_palindrome_slicing("Hello"))  # False
````

````python
# 2nd method
def is_palindrome_two_pointer(s):
    s = s.lower().replace(" ", "")  # Optional: ignore case and spaces
    left = 0
    right = len(s) - 1
    
    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1
    
    return True

# Examples
print(is_palindrome_two_pointer("racecar"))        # True
print(is_palindrome_two_pointer("Hello"))          # False
print(is_palindrome_two_pointer("A man a plan a canal Panama"))  # True
````

````python
# Using a Loop (Iterative Approach)
def is_palindrome_loop(s):
    s = s.lower().replace(" ", "")
    n = len(s)
    for i in range(n // 2):
        if s[i] != s[n - i - 1]:
            return False
    return True

# Example
print(is_palindrome_loop("racecar"))  # True
````

````python
# Print all the subsequence 


arr = [3, 1, 2]
subsequences = [[]]  # start with an empty subsequence

for num in arr:
    # for each element, add it to all existing subsequences
    new_subseqs = []
    for seq in subsequences:
        new_subseqs.append(seq + [num])
    subsequences.extend(new_subseqs)

# print all subsequences
for seq in subsequences:
    print(seq)
````
