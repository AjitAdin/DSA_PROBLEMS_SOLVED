```python
# Re 3. Parameterised and Functional Recursion 

# sum of n natural numbers

# Parameterized Recursion

def fun(i, sum):
    if i < 1:            # base case
        print(sum)       # print final result
        return
    fun(i - 1, sum + i)  # recursive call with updated sum

fun(5, 0)
```

```python
# functional
def fun(n):
    if n == 0:        # base case
        return 0
    return n + fun(n - 1)   # recursive return

print(fun(5))
```

# Problems on Functional Recursion 

```python
arr = [1, 2, 3, 4, 2]
l = 0
r = len(arr) - 1

while l < r:
    arr[l], arr[r] = arr[r], arr[l]  # swap elements
    l += 1
    r -= 1

print(arr)
```

```python
# parametrized 
def reverse(arr,l,r):
    if l>=r:
        return
    arr[l], arr[r] = arr[r], arr[l]
    
    reverse(arr,l+1,r-1)

arr=[1,2,3,4,2]
reverse(arr,0,len(arr)-1)
print(arr)
```

```python
# check if the string is palindrome or not
def is_palindrome_slicing(s):
    s = s.lower().replace(" ", "")  # Optional: ignore case and spaces
    return s == s[::-1]

# Example
print(is_palindrome_slicing("Madam"))  # True
print(is_palindrome_slicing("Hello"))  # False
```

```python
# 2nd method
def is_palindrome_two_pointer(s):
    s = s.lower().replace(" ", "")  # Optional: ignore case and spaces
    left = 0
    right = len(s) - 1
    
    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1
    
    return True

# Examples
print(is_palindrome_two_pointer("racecar"))        # True
print(is_palindrome_two_pointer("Hello"))          # False
print(is_palindrome_two_pointer("A man a plan a canal Panama"))  # True
```

```python
# Using a Loop (Iterative Approach)
def is_palindrome_loop(s):
    s = s.lower().replace(" ", "")
    n = len(s)
    for i in range(n // 2):
        if s[i] != s[n - i - 1]:
            return False
    return True

# Example
print(is_palindrome_loop("racecar"))  # True
```

```python
# recursive approach


def fun(i,s):
    if i>=len(s)//2:
        return True
    
    if s[i] != s[len(s) - i - 1]:
        return False
    return fun(i+1,s)
    



s="madam"
print(fun(0,s))
```

```python
def fibonacci(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    else:
        f,s=0,1
        for i in range(n-1):
            f,s=s,f+s
        return s

print(fibonacci(8))

# def fibonacci(n):
#     if n == 0:
#         return 0
#     f, s = 0, 1
#     for _ in range(1, n):
#         f, s = s, f + s
#     return s

# # Example:
# print(fibonacci(8))  # Output: 21
```

```python
# recursive approach

def fibonacci(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    else:
        return fibonacci(n-2)+fibonacci(n-1)


print(fibonacci(8))


# or

# def fibonacci(n):
    
#     if n<=1:
#         return n
#     else:
#         return fibonacci(n-2)+fibonacci(n-1)


# print(fibonacci(8))


# | Aspect               | Complexity                      |
# | -------------------- | ------------------------------- |
# | **Time Complexity**  | `O(2^n)` (exponential)          |
# | **Space Complexity** | `O(n)` (due to recursion stack) |
```

```python
# Print all the subsequence 


arr = [3, 1, 2]
subsequences = [[]]  # start with an empty subsequence

for num in arr:
    # for each element, add it to all existing subsequences
    new_subseqs = []
    for seq in subsequences:
        new_subseqs.append(seq + [num])
    subsequences.extend(new_subseqs)

# print all subsequences
for seq in subsequences:
    print(seq)
```
