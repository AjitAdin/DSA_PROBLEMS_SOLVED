```python
def solve(i, arr, f):
    if i == len(arr):
        print(f)
        return
    # pick the element
    solve(i + 1, arr, f + [arr[i]])
    # don't pick the element
    solve(i + 1, arr, f)


def main():
    arr = [3, 1, 2]
    print("All possible subsequences are:")
    solve(0, arr, [])


if __name__ == "__main__":
    main()
```

```python
# beeter approach

def solve(i, arr, res, k, current_sum):
    # base case
    if i == len(arr):
        if current_sum == k:
            print(res)
        return
    
    # pick the element
    solve(i + 1, arr, res + [arr[i]], k, current_sum + arr[i])
    
    # don't pick the element
    solve(i + 1, arr, res, k, current_sum)


arr = [1, 2, 3]
k = 3
solve(0, arr, [], k, 0)
```

```python
# printing the subsequence any one only
def solve(i, arr, res, k):
    if sum(res) == k:  # check sum at any point
        print(res)
        return True   # stop recursion after finding one subset
    if i == len(arr):
        return False

    # pick the element
    if solve(i + 1, arr, res + [arr[i]], k):
        return True

    # don't pick the element
    if solve(i + 1, arr, res, k):
        return True

    return False

arr = [1, 2, 3]
k = 3
solve(0, arr, [], k)


```

```python
# COUNT THE SUBSEQUENCE WITH SUM==k

def count_subseq(i, arr, res, k):
    if i == len(arr):
        if sum(res) == k:
            return 1
        else:
            return 0
    # Include arr[i]
    count_incl = count_subseq(i + 1, arr, res + [arr[i]], k)
    # Exclude arr[i]
    count_excl = count_subseq(i + 1, arr, res, k)
    return count_incl + count_excl

arr = [1, 2, 1,1,3]
k = 3
print(count_subseq(0, arr, [], k))  # Output: 2
```

```python
# Inefficiency: You are creating a new list res + [arr[i]] in every recursive call, which is unnecessary. We only care about the sum, not the actual subsequence.

# Optimization: Instead of storing the subsequence, we can just keep track of the current sum. This reduces both time and space usage.

# Hereâ€™s an improved version:

def count_subseq(i, arr, curr_sum, k):
    if i == len(arr):
        return 1 if curr_sum == k else 0
    
    # Include arr[i]
    count_incl = count_subseq(i + 1, arr, curr_sum + arr[i], k)
    # Exclude arr[i]
    count_excl = count_subseq(i + 1, arr, curr_sum, k)
    
    return count_incl + count_excl

arr = [1, 2, 3]
k = 3
print(count_subseq(0, arr, 0, k))  # Output: 2

```
