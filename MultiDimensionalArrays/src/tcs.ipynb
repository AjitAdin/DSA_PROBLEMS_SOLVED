```python
# To convert a given matrix into a lower triangular matrix, you need to set all the elements above the main diagonal to zero. The main diagonal consists of elements where the row index equals the column index (i.e., matrix[i][i]).

# Example:

# If the input matrix is:

# 1234
# 5678
# 9876
# 5432


# After converting it into a lower triangular matrix, it will look like this:

# 1 0 0 0
# 5 6 0 0
# 9 8 7 0
# 5 4 3 2




arr = [
    [1, 2, 3, 4],
    [5, 6, 7, 8],
    [9, 8, 7, 6],
    [5, 4, 3, 2]
]

row = len(arr)

# Convert the matrix to a lower triangular matrix
for i in range(row):
    for j in range(i + 1, row):  # For each element above the main diagonal (i < j)
        arr[i][j] = 0

# Print the resulting lower triangular matrix
for i in range(row):
    print(arr[i])
```

```python
def lower_triangular(matrix):
    """
    Converts a square matrix to a lower triangular matrix.

    Args:
        matrix: A square matrix represented as a list of lists.

    Returns:
        A new list of lists representing the lower triangular matrix.
    """
    rows = len(matrix)
    cols = len(matrix[0])
    
    # Check if the matrix is square
    if rows != cols:
        print("Matrix is not square.")
        return None

    # Create a new matrix filled with zeros
    result = [[0] * cols for _ in range(rows)]

    # Iterate through the matrix and copy elements on and below the diagonal
    for i in range(rows):
        for j in range(cols):
            if i >= j:
                result[i][j] = matrix[i][j]

    return result

# Example usage with the matrix from the image
input_matrix = [
    [1, 2, 3, 4],
    [5, 6, 7, 8],
    [9, 8, 7, 6],
    [5, 4, 3, 2]
]

lower_triangular_matrix = lower_triangular(input_matrix)

if lower_triangular_matrix:
    for row in lower_triangular_matrix:
        print(row)
```

```python
import math

# Calculate the square root
sqrt_value = math.sqrt(12)

# Find the floor (lower bound) and ceiling (upper bound)
lower_bound = math.floor(sqrt_value)
upper_bound = math.ceil(sqrt_value)

print(f"Square root of 12: {sqrt_value}")
print(f"Lower bound: {lower_bound}")
print(f"Upper bound: {upper_bound}")
```
